```
import imageio as im
```

```
img = im.imread('sample.jpg')
print(img.shape)
print(img.dtype)
print(type(img))
```

```
import matplotlib.pyplot as plt
```

```
plt.imshow(img)
plt.show()
```

```
import numpy as np
np.random.seed(0)
```

```
noise = np.random.normal(0, 5, size=img.shape)
#noise
```

```
noisy_img = img + noise
plt.imshow(noisy_img.astype(np.uint8))
plt.show()
```

```
fliped_img = np.fliplr(noisy_img)
plt.imshow(fliped_img.astype(np.uint8))
plt.show()
```

```
rotated_img = np.rot90(noisy_img)
plt.imshow(rotated_img.astype(np.uint8))
plt.show()
```

```
hist , bin_edges = np . histogram (img)
```

```
hist, bin_edges = np.histogram(img[:,:,0], bins=20)
width = 0.8 * (bin_edges[1] - bin_edges[0])
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
plt.bar(bin_centers, hist, width=width)
plt.title('Red') 
plt.ylabel('Count')
plt.xlabel('Intensity')
plt.show()
```
